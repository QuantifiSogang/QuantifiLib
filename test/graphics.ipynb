```python
from quantifilib.data import YahooFinanceLoader

loader = YahooFinanceLoader()
data = loader.load(
    'SPY',
    start = '2000-01-01',
    interval = '1d',
    auto_adjust = False,
    multi_level_index = False,
    progress = False
)
```
